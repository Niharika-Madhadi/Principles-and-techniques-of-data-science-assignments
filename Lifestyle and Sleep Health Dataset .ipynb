 The Lifestyle and Sleep Health Dataset comprises 400 rows and 13 columns, covering a wide range of variables related to sleep and daily habits.

 The attributes are :

**Person ID**: It is an identifier given to each individual.

**Gender**: The gender of the individual(Male/Female)

**Age**: The age of the person in years.

**Occupation**: The profession of the person.

**Sleep Duration** (hours): The number of hours the person sleeps per day.

**Quality of Sleep** (scale: 1-10): A subjective rating of the quality of sleep, ranging from 1 to 10.

**Physical Activity Level** (minutes/day): The number of minutes the person engages in physical activity daily.

**Stress Level** (scale: 1-10): A subjective rating of the stress level experienced by the person, ranging from 1 to 10.

**BMI Category**: The BMI category of the person (e.g., Underweight, Normal, Overweight).

**Blood Pressure** (systolic/diastolic): The blood pressure measurement of the person, indicated as systolic pressure over diastolic pressure. Which should be less than 120/80.

**Heart Rate** (bpm): The resting heart rate of the person in beats per minute. Which is normally from 60-100.

**Daily Steps**: The number of steps the person takes per day.

**Sleep Disorder**: The presence or absence of a sleep disorder in the person (None, Insomnia, Sleep Apnea).

None: The individual does not exhibit any specific sleep disorder.

Insomnia: The individual experiences difficulty falling asleep or staying asleep, leading to inadequate or poor-quality sleep.

Sleep Apnea: The individual suffers from pauses in breathing during sleep, resulting in disrupted sleep patterns and potential health risks.

The dataset is used to suggest the sleep health and lifestyle practices that will reduce the health risks. It has details of Sleep Metrics like Explore sleep duration, quality, and factors influencing sleep patterns.It features the Lifestyle Factors like Analyze physical activity levels, stress levels, and BMI categories. It gives ide about our Cardiovascular Health by Examining blood pressure and heart rate measurements.It tells about the Sleep Disorder Analysis by Identifying the occurrence of sleep disorders such as Insomnia and Sleep Apnea.
