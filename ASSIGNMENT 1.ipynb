## Measures of Centre

We have found the measures of centre(mean, median, mode) of age feature in titanic dataset.

We can infer that average of age of people present in titanic is approximately 30. When ordered and got middle value as 28, we can infer that half of the people comes below age of 28. By using value_counts() function we can see that there are 30 young people who are aged 24 years and the oldest persons age is 74 years. From mode we can see that mostly repeated people are of 24 years old.

```python
import numpy as np

print('Mean :',np.mean(Titanic['age']))
print('Median :', Titanic['age'].median())
print(Titanic['age'].value_counts())
print('Mode :',Titanic['age'].mode())
```

We are plotting mean and median of age feature and can see that mean>median

```python
# plotting mean, median
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

xmean = np.mean(Titanic['age'])
xmedian = (Titanic['age']).median()

fig = plt.figure(figsize=(10,5))
sns.histplot(Titanic['age'], kde=True, color='grey')
plt.axvline(x=xmean, color='red', label='Mean', linewidth=7)
plt.axvline(x=xmedian, color='green', label='Median', linewidth=5)
plt.legend()
plt.show()
```

## Measures of Spread

We found the measures of spread of survived feature by calculation the variance and standard deviation of it using var() and std() functions in numpy. We plot the normal distribution curve of survived feature

```python
import numpy as np
import matplotlib.pyplot as plt

print('Variance :', np.var(Titanic['survived']))
print('Standard deviation :', np.std(Titanic['survived']))
x_axis=Titanic['survived']
mu=np.mean(Titanic['survived'])
sigma=np.std(Titanic['survived'])
x = np.random.normal(0, 1, 1000)
count, bins, ignored = plt.hist(x, density=True, bins=25)
plt.plot(bins, 1/(np.sqrt(2 * np.pi * sigma**2)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=3, color='k')
plt.xlabel('Standard Deviations')
plt.ylabel('Probabilities')
plt.axvline(x=mu, color='red', label='Mean')
plt.title('Normal Standard Distribution')
plt.legend()
plt.show()
```

## Measures of Shape

We know that mean>median for age feature from above and here when we calculate Skewness we got a positive number, hence the data is right-skewed.

The Kurtosis is 0.18 which is less than 3, hence it is platykurtic

```python
print('Kurtosis:',Titanic['age'].kurtosis())
print('Skewness :',Titanic['age'].skew())
```
