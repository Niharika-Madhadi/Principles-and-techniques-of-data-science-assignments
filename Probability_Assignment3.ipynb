1.A fair six-sided die is rolled twice. What is the probability of getting 2 on the first roll and not getting 4 on the second roll?

P(Getting 2 on first roll)= $\frac{1}{6}$

P(not getting 4 on second roll) = $\frac{5}{6}$

As the two events are independent of each other, we can multiply both probabilities.

$P(A\cap B)= P(A)*P(B)$ = $\frac{1}{6}*\frac{5}{6}$=$\frac{5}{36}$

2.“Let’s say you have 2 dice. What is the probability of getting at least one 4?”

This question is one of the examples of independent events since the outcome from throwing a die wouldn’t have any effect on the outcome from throwing the second die.

Let’s say that:

A = getting a 4 in the first die

B = getting a 4 in the second die

The probability of independent events A and B both to occur can be defined
$P(A\cap B)= P(A)*P(B)$

And the probability of getting at least one 4 can be computed with the probability of union of two events:

$P(A \cup B)=P(A)+P(B)-P(A\cap B)$

We know that the probability of us getting any specific outcome from throwing a die is ⅙. Thus,

$P(A \cup B)=\frac{1}{6}+\frac{1}{6}-(\frac{1}{6}*\frac{1}{6})= \frac{11}{36}$


3.Hospital records show that 75% of patients suffering from a disease die due to that disease. What is the probability that 4 out of the 6 randomly selected patients recover?

Think of this as a binomial since there are only 2 outcomes, either the patient dies or he survives.

Here n =6, and x=4.

p=0.25(probability if living(success))

q = 0.75(probability of dying(failure))

P(X) = $C_x^{n}*p*x*q*(n-x)$ = $C(6,4)*0.25*4*0.75*2$ = 0.03295

We will solve the above problem using python. As it is binomial distribution we use binom function from stats package. As the probability here is equal to some particular value we use the pmf() function to find the value.

```python
import scipy.stats as stats

stats.binom.pmf(4,6,0.25)
```

We will plot the pmf graph for above question.

```python
# probability mass function
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
# binom.pmf(x, n, p, loc)

n = 6
p = 0.25
chances = list(range(1, n + 1))
probs = [stats.binom.pmf(x, n, p) for x in chances]

plt.bar(chances, probs)
plt.xlabel('Chances')
plt.ylabel('Probability')
plt.show()
```

We can see that in entire sample space of 1-6, the probability that those members will survive is shown. We can infer that probablity that one person will survive out of 6 is highest (0.35) and the probability that 4 persons will survive is 0.035 as we solved above. The probability that all 6 persons will survive is the least. As it is discrete distribution, the sum of all the probabilities in the graph will be equal to 1.

4."What percentage of results have 0.2 or fewer heads?"

Since we cannot get a partial number of heads, but it's easy to calculate that the only number of heads that is equal to or less than 0.2 is getting zero heads. Since this only happens one our of four times, we expect the result to be 25%.

```python
from scipy.stats import binom

Coin= binom(n=2,p=0.5)
res=Coin.cdf(0.2)
print('The result is ',res)
```

```python
# plot the cdf
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

x = np.linspace(-4, 4, 1000)
cdf_y = stats.norm.cdf(x)
plt.plot(x, cdf_y, 'b', label='cdf')
plt.title('The CDF')
plt.xlabel('X')
plt.ylabel('cdf(X)')
plt.grid(True)
plt.show();
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the probability density function
def pdf(x):
  return np.exp(-x**2 / 2) / np.sqrt(2 * np.pi)

# Generate some data
x = np.linspace(-3, 3, 100)
y = pdf(x)

# Plot the probability density function
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("Probability density")
plt.title("Probability density function of the normal distribution")
plt.show()
```

We have defined the pdf using it's mathematical formula. We then generated the data points using linspace() function and used pdf() function to get the results for those data points.

5.How do you calculate the probability of drawing a red card and then face card from a standard deck of cards without replacement?

There are 26 red cards in a standard deck of cards, and there are 52 cards in total. There are 12 face cards from the deck, but this time the total number of cards is 51.

The probability of drawing a red card P(A)= $\frac{26}{52}$ =$\frac{1}{2}$.

The probability of drawing face card P(B|A) = $\frac{12}{51}$

$P(A\cap B)=P(A)*P(B|A)$ = $\frac{1}{2}$*$\frac{12}{51}$ = $\frac{6}{51}$
